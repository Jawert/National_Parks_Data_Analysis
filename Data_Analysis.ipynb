```python
# Standard data science imports
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import seaborn as sns

# Import our cleaning script
import data_clean
```

```python
trails_info = data_clean.cleanit()
```

```python
# Quick look at the data
trails_info.head()
```

```python
# Creating a scatterplot to see if there is any relationship between geographical location and difficulty of the trails

import matplotlib.pyplot as plt
sns.set_style('whitegrid')

# Create teh figure and set the size
fig, scatter = plt.subplots(figsize = (12.5,8.5))

# Create a scatterplot of longitude vs. latitude, with the size of the dots as the popularity of the trail, and the 
# color of the dots determined by the difficulty_rating.
scatter = sns.scatterplot(x = 'lon', y ='lat', data=trails_info, size = 'popularity', hue = 'difficulty_rating', palette = 'muted')

# Set the axis labels
scatter.set_xlabel('Longitude', fontsize = 18)
scatter.set_ylabel ('Latitude', fontsize = 18)
```

```python
# Creating a scatterplot to see if there is any relationship between geographical location and route_type of the trails

import matplotlib.pyplot as plt
sns.set_style('whitegrid')

# Create teh figure and set the size
fig, scatter = plt.subplots(figsize = (12.5,8.5))

# Create a scatterplot of longitude vs. latitude, with the size of the dots as the popularity of the trail, and the 
# color of the dots determined by the route_type.
scatter = sns.scatterplot(x = 'lon', y ='lat', data=trails_info, size = 'popularity', hue = 'route_type', palette = 'muted')

# Set the axis labels
scatter.set_xlabel('Longitude', fontsize = 18)
scatter.set_ylabel ('Latitude', fontsize = 18)
```

```python
# Set the figure size
fig, hist = plt.subplots(figsize = (12.5,8.5))

# Create a stacked barchart of popularity and route_type
hist = sns.histplot(data = trails_info, x = 'popularity', multiple = 'stack', hue = 'route_type', bins = 75)
```
